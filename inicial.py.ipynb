**¿Qué son los AG?**

Son métodos adaptativos que pueden usarse para resolver problemas de búsqueda y optimización, que inspiran en 
la teoría de evolución natural.  

Se llama *población* al conjunto de soluciones e *individuo* a cada una de las soluciones; mientras que otros 
autores al conjunto de soluciones lo dominan *fenotipo* mientras que a la información que determina a cada individuo 
*cromosoma*.

El algoritmo evalúa las soluciones y se selecciona la que mejor resulva el problema.

Para cuantificar la distancia entre el valor deseado **target** y el valor incial **source** vamos a usar nuestra la 
funcion **fitness**, la comparación es caracter a caracter, cada uno de estos caracteres se convierten a su correspondiente
valor ascii y se mide la distacion entre ellos.  
La condición física perfecta es 0. 

```python
def fitness(source, target):
    fitval = 0
    for i in range(0, len(source)):
        fitval += (ord(target[i]) - ord(source[i])) ** 2
    return(fitval)

def fitnesspri(source, target):
    fitval = 0
    print(target,source)
    for i in range(0, len(source)):
        fitval += (ord(target[i]) - ord(source[i])) ** 2
        #print(i,ord(target[i]),ord(source[i]),fitval)   
    return(fitnesspri)
```

Vamos a usar 2 funciones mutate

```python
def mutate(source):
    charpos = random.randint(0, len(source) - 1)
    parts = list(source)
    parts[charpos] = chr(ord(parts[charpos]) + random.randint(-1,1))
    return(''.join(parts))

def getMutate(parent):
    geneIndex = random.randint(0, len(geneset) -1);
    index = random.randint(0, len(parent) - 1)
    genes = list(parent)
    genes[index] = geneset[geneIndex]
    return(''.join(genes))
```

La funcion **mutate** selecciona un caracter al azar, al valor de este caracter se incrementa o decrementa en uno su valor o se queda igual.

```python
fitval = fitness(source, target)
#print("fitval %d" %(fitval))
i = 0
while True:
    i += 1
    m = mutate(source)
    fitval_m = fitness(m, target)
    if fitval_m < fitval:
        fitval = fitval_m
        source = m
        print ("%5i %5i %14s" % (i, fitval_m, m))
    if fitval == 0:
        break
```
