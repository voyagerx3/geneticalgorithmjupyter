# Algoritmos Genéticos

#### La mejor tapa de Almería un Shérigan 

**¿Qué son?**

se inspiran en la evolución natural para solucionar problemas de optimización, que de otra forma serían difíciles para un diseñador humano.
Se llama población al conjunto de soluciones e individuo a cada una de las soluciones.
El algoritmo evalúa las soluciones y se selecciona la que mejor resulva el problema.

**Descripción del problema:**
    
En este algoritmo vamos a usar un string de  36 caracteres y a través de mutaciones
se busca llegar a la frase __target__ **"La mejor tapa de Almería un Shérigan"** 



```python

import random
```

```python
target = "La mejor tapa de Almería un Shérigan"

print(len(target))
```

```python
char = "abcdefghijklmnñopqrstuvwyzABCDEFGHIJKLMÑOPQRSTÁÉÍÓÚáéíóú=)(\¿?¡!"
nword=len(target)
def disorder(char,nword):
    tword = []
    seperator = ', '
    for i in range(1, nword+1):
        letra_aleatoria = random.choice(char)
        tword.append(letra_aleatoria)
    return "".join(str(x) for x in tword)

source= disorder(char,nword) 
print("source = \"{}\"".format(source))

```

```python
def fitness(source, target):
    fitval = 0
    for i in range(0, len(source)):
        fitval += (ord(target[i]) - ord(source[i])) ** 2
    return(fitval)
```

```python
def mutate(source):
    charpos = random.randint(0, len(source) - 1)
    parts = list(source)
    parts[charpos] = chr(ord(parts[charpos]) + random.randint(-1,1))
    return(''.join(parts))
```

```python
fitval = fitness(source, target)
i = 0
while True:
    i += 1
    m = mutate(source)
    fitval_m = fitness(m, target)
    if fitval_m < fitval:
        fitval = fitval_m
        source = m
        print ("%5i %5i %14s" % (i, fitval_m, m))
    if fitval == 0:
        break
```
